**Q7**. While Loop
Write a program that prints numbers from 1 to 50 but:

- Skip all numbers divisible by 4

- Stop completely when number reaches 37

```python

i = 1               # start with 1
while i <= 50 :
  if i == 37:       # If the number reaches 37, stop the loop completely
        break

  if i % 4 == 0:    # If number is divisible by 4, skip this number and move to the next
        i += 1      # increment before continue
        continue    # skip print the number, those are devided by 4
  print(i)
  i += 1            # increment by 1 by 1




```

# Q8. For Loop

Write a program that prints the multiplication table of a user-entered number from 1 to 12.

```python
# Take number from user
num = int(input("Enter a number: "))

# Print multiplication table from 1 to 12
for i in range(1, 13):
    print(num, "x", i, "=", num * i)

```

 Write a program that:

**Q12. List Comprehension**

Using list comprehension, create a list of all numbers from 1 to 100 that are:

- Divisible by 3

- But not divisible by 5



```python
# task 5

nums = [n for n in range(1, 101) if n % 3 == 0 and n % 5 != 0]      # Select only the numbers divisible by 3 AND not divisible by 5

print(nums)
```
